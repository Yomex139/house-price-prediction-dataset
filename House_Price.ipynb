```python
def Test(df1):
    for column,contents in df1.items():
        if pd.api.types.is_numeric_dtype(contents):
            if contents.isnull().sum():
                df1[column +'_is_missing'] = pd.isnull(contents)
                df1[column] = contents.fillna(contents.median())
                for column,contents in df1.items():
                    if pd.api.types.is_string_dtype(contents):
                        df1[column] = pd.Categorical(contents).codes +1
    return df1
```
